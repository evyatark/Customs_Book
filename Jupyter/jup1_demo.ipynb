```python
df1 = pd.read_excel("./טבלה מרכזית.xlsx", sheet_name=1)
df1 = df1.rename(columns={"פרט מכס":"Custom_Item"})
df1["Custom_Item"] = df1["Custom_Item"].astype(object)
# above line causes a warning. we should use this instead: df1 = df1.astype({"Custom_Item":str}, errors='raise')
df1 = df1.set_index("Custom_Item")

df2 = pd.read_excel("./טבלה מרכזית.xlsx", sheet_name=2)
df2 = df2.rename(columns={"פרט מכס":"Custom_Item"})
df2["Custom_Item"] = df2["Custom_Item"].astype(object)
df2 = df2.set_index("Custom_Item")

df3 = pd.read_excel("./טבלה מרכזית.xlsx", sheet_name=3)
df3 = df3.rename(columns={"פרט":"Custom_Item"})
df3["Custom_Item"] = df3["Custom_Item"].astype(object)
df3 = df3.set_index("Custom_Item")

# IMPORTANT! note that we start with df3! because it has more rows than the 2 other data-frames
# (this might be incorrect in future when we receive a new Excel workbook with different data!)
# also - we assume that all values in index of df1, df2 already exist in the index of df3!! 
# (using merge() should correct this problem)
df = df3

# add columns from other data-frames (= worksheets)
df["כמות היבואנים עם זיהוי"] = df1["כמות היבואנים עם זיהוי"]
df["כמות סוכנים"] = df2["כמות סוכנים"]
#df["ספירת הצהרות"] = df3["ספירת הצהרות"]

# to see that the merging of the 3 worksheets was done correctly
# display the last 40 lines, and check that values of each custom item where copied correctly
df.tail(40)

#df.info()

# Note: I think it would be better to use data-frame merge() function - 
# it should use the custom_item index correctly.


```

```python
# .astype(object) together with set_index() causes the warnings (FutureWarning: In a future version, the Index constructor will not infer numeric dtypes when passed object-dtype sequences (matching Series behavior))
# to eliminate the warnings, instead of this:
#df1["Custom_Item"] = df1["Custom_Item"].astype(object)
# use this:
#df1 = df1.astype({"Custom_Item":str}, errors='raise')

df1 = pd.read_excel("./טבלה מרכזית.xlsx", sheet_name=1)
df1 = df1.rename(columns={"פרט מכס":"Custom_Item"})
#df1["Custom_Item"] = df1["Custom_Item"].astype(object)
df1 = df1.astype({"Custom_Item":str}, errors='raise')
df1 = df1.set_index("Custom_Item")
```

```python
df1 = pd.read_excel("./טבלה מרכזית.xlsx", sheet_name=1)
df1 = df1.rename(columns={"פרט מכס":"Custom_Item"})
df1 = df1.astype({"Custom_Item":str}, errors='raise')
df1 = df1.set_index("Custom_Item")

df2 = pd.read_excel("./טבלה מרכזית.xlsx", sheet_name=2)
df2 = df2.rename(columns={"פרט מכס":"Custom_Item"})
df2 = df2.astype({"Custom_Item":str}, errors='raise')
df2 = df2.set_index("Custom_Item")

df3 = pd.read_excel("./טבלה מרכזית.xlsx", sheet_name=3)
df3 = df3.rename(columns={"פרט":"Custom_Item"})
df3 = df3.astype({"Custom_Item":str}, errors='raise')
df3 = df3.set_index("Custom_Item")

# Here also the order of merging matters, we must start with df3 (because we use left-join!!)
#df = df3.copy()
#df = df.merge(df2, how="left", left_on="Custom_Item", right_index=True)
#df.dtypes
#df = df.merge(df1, how="left", left_on="Custom_Item", right_index=True)
#df.tail(40)

# Using outer join - This will work in any orderof merges!!
df = df1.copy()
df = df.merge(df2, how="outer", left_on="Custom_Item", right_on="Custom_Item", indicator=False)
df = df.merge(df3, how="outer", left_on="Custom_Item", right_on="Custom_Item", indicator=False)
df = df.sort_index()
df.tail(40)
#df.info()
```
